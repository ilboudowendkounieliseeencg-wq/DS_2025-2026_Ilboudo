```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/drive/MyDrive/DS_Elisee Ilboudo/xAPI-Edu-Data.csv'
df = pd.read_csv(file_path)
display(df.head()) # variable.head permet d'afficher les 5 premières lignes du dataset
                  # Pour voir tout le dataframe il suffit d'appeler la variable
```

```python
# Statistiques descriptives
print(df.describe())

```

```python
# Statistiques pour chaque colonne numérique
numerical_cols = df.select_dtypes(include=np.number).columns

for colonne_name in numerical_cols:
  colonne = df[colonne_name]
  print(f"Statistiques pour la colonne: {colonne_name}")
  print(f"Moyenne: {colonne.mean():.2f}")
  print(f"Médiane: {colonne.median():.2f}")
  # Mode might have multiple values, so handling it differently
  mode_values = colonne.mode()
  print(f"Mode: {', '.join([f'{m:.2f}' for m in mode_values])}")
  print(f"Écart-type: {colonne.std():.2f}")
  print(f"Variance: {colonne.var():.2f}")
  print(f"Min: {colonne.min():.2f}")
  print(f"Max: {colonne.max():.2f}")
  print("-" * 30) # Separator for better readability
```

```python
# Quartiles pour chaque colonne numérique
numerical_cols = df.select_dtypes(include=np.number).columns

for colonne_name in numerical_cols:
  colonne = df[colonne_name]
  print(f"Quartiles pour la colonne: {colonne_name}")
  print(f"Q1 (25%): {colonne.quantile(0.25):.2f}") # le ".2f" signifie qu'il faut s'arreter au 2eme decimal
  print(f"Q2 (50% - Médiane): {colonne.quantile(0.50):.2f}")
  print(f"Q3 (75%): {colonne.quantile(0.75):.2f}")
  print(f"IQR: {colonne.quantile(0.75) - colonne.quantile(0.25):.2f}")
  print("-" * 30) # Separator for better readability
```
